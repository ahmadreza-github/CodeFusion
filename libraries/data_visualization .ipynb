pie charts

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv(r"C:\Users\AHMAD\OneDrive\Desktop\program(github)\datasets\diabetes.csv")
youngers = data.loc[data['Age'] <= 35].count()[0]
olders = data.loc[data['Age'] <= 40].count()[0]       
labels = ["youngers" ,"olders"] 
colors = ['blue' , "red"]
plt.pie([youngers , olders] , labels = labels , colors = colors , autopct = '%.2f %%')   
# '%%puts the symbol of percent in front of the numbers'
# '%.2f' gets the percent and prints it.
plt.show() 

```

a project with piechart 

```python
# data.info()
# data.BMI.describe()
# 22.3
# min BMI is 0 
# max BMI is 67
# noraml BMI is 32
low_bmi = data.loc[(data['BMI'] >= 0) & (data["BMI"] <= 22)].count()[0]

medium_bmi = data.loc[(data["BMI"] >= 23) & (data["BMI"]<= 44)].count()[0]

high_bmi = data.loc[(data['BMI'] >= 45) & (data["BMI"] <= 68)].count()[0] 

labels = ["low bmi" , "medium_bmi" , "high bmi"]

colors = ['purple' , "blue" , "orange"]

explode = (.9 , .1 , .5) 

plt.pie([low_bmi , medium_bmi , high_bmi] , labels = labels , autopct = '%.2f %%' , pctdistance = 0.8, colors = colors , explode = explode)
# pctdistance : you can determine the distant of 'autopct'
plt.style.use('default') 
plt.title('Different BIMs for different ages')
plt.show()  
# you can access different styles to use by the command: (' print(plt.style.available) ') 
```

boxplot

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\AHMAD\OneDrive\Desktop\program(github)\datasets\gas_prices.csv") 
plt.figure(figsize = (5, 8)) 
x = data["France"]
y = data["Japan"]
z = data["Canada"]
labels = ["France" , "Japan" ,"canada"]

boxes = plt.boxplot([x , y , z], labels = labels , patch_artist = True)
for box in boxes['boxes']:
    box.set(color = "red" , linewidth = 3)
    box.set(facecolor = "purple") # the color inside the square will chane
    # patch_artist = True is required

plt.ylabel("numbers")
plt.xlabel("countries")
```
