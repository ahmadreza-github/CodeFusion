```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr
# a simple model to guess an algorithm:
# -1 means: get the value required by yourself
x = np.array([1, 2, 3, 4]).reshape(-1, 1)
y = np.array([1, 3, 5, 7])
plt.plot(x, y, "o")
x.shape

model = lr()
model.fit(x, y)
m = model.coef_[0]
c = model.intercept_
result = model.predict([[10]])
print(result)

# new model training for guessing a house pirce based on its square feet
# Data preparation
size = np.array([500, 1000, 1500, 2000, 2500]).reshape(-1, 1)
prices = np.array([150, 200, 250, 300, 350]).reshape(-1, 1)

# Scatter plot
plt.scatter(size, prices, color="blue")
plt.title("House Prices vs Size")
plt.xlabel("Size (square feet)")
plt.ylabel("Price (thousands of dollars)")
plt.grid()
plt.show()

# Model fitting
model = LinearRegression()
model.fit(size, prices)  # Corrected input order

# Extracting slope and intercept
slope = model.coef_[0][0]  # Slope (m)
intercept = model.intercept_[0]  # Intercept (c)

# Prediction
predicted_price = model.predict([[1800]])[0][0]  # Get the predicted price

# Output results
print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
print(f"Predicted price for a house of size 1800 sq ft: ${
      predicted_price:.2f} thousand")
# (:.2f) is a sort of float formatting
# ( : )  the start of the formatting specification.
# (.2) : specifies that the number should be rounded and displayed with exactly two digits
# (f) : indicates that the number is a floating-point number.
```

guessing a car's price based on model:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression as lr

camery_model = np.array([2010, 2012, 2014, 2016, 2018, 2020]).reshape(-1, 1)
camery_price = np.array(
    [40000, 60000, 80000, 100000, 120000, 140000]).reshape(-1, 1)

plt.plot(camery_model, camery_price, marker="*")
plt.title("Toyota Camery prices(year of production)")
plt.ylabel("prices")
plt.xlabel("year of production")
plt.grid()
model = lr()
model.fit(camery_model, camery_price)
m = model.coef_[0]
c = model.intercept_
result = model.predict([[2022]])
# print(result)
print(f"a toyota camery with the model 2022 can cost you: {result}")
```

```python
# model using DecisionTreeRegressor
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression as lr
experience_years = np.array([0 , 1 , 2 , 3 , 4 , 6 , 8 , 10 ]).reshape(-1 , 1)
salaries = np.array([30000, 40000, 55000, 60000, 70000, 80000, 85000, 87000]).reshape(-1, 1)

# data visualization:
plt.scatter(experience_years , salaries ) 

# data fiting
model = DecisionTreeRegressor(max_depth=4)

model.fit(experience_years , salaries)

y_predict= model.predict(experience_years)

plt.plot(experience_years , salaries , "o")
plt.plot(experience_years , y_predict)
plt.show()
```
